## Importing Libraries

```python
# Import necessary libraries
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
import cv2
```

## Task 2: Load and Explore Data

```python
# Load a pre-trained model (e.g., MobileNetV2)
model = tf.keras.applications.MobileNetV2(weights='imagenet')
```

## Task 3: Data Preprocessing and Normalization

```python
# Load and preprocess the image (image3.jpg)
img_path = 'image3.jpg'  # replace with your image file path
img = image.load_img(img_path, target_size=(224, 224))  # Resize to fit model input size
img_array = image.img_to_array(img)  # Convert the image to an array
img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
img_array = tf.keras.applications.mobilenet_v2.preprocess_input(img_array)  # Preprocess for MobileNetV2
```

## Task 4: Build, Train, and Evaluate the Model

```python
# Make predictions
predictions = model.predict(img_array)
# Decode predictions to get human-readable labels
decoded_predictions = tf.keras.applications.mobilenet_v2.decode_predictions(predictions, top=3)[0]
# Decode predictions to get human-readable labels
decoded_predictions = tf.keras.applications.mobilenet_v2.decode_predictions(predictions, top=3)[0]
```

## Task 5: Test the Model with Custom Images

```python
# Display the image and the top 3 predictions
plt.imshow(img)
plt.axis('off')  # Hide axes
plt.show()
# Print the predictions
for i, (imagenet_id, label, score) in enumerate(decoded_predictions):
    print(f"{i + 1}. {label}: {score:.2f}")
```
